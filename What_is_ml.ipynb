# 1.머신러닝이란
머신러닝은 데이터에서부터 학습하도록 컴퓨터를 프로그래밍하는 과학이다.

톰 미첼에 따르면 어떤 작업 T에 대한 컴퓨터 프로그램의 성능을 P로 측정했을 때 경험 E로 인해 성능이 향상됐다면, 이 컴퓨터 프로그램은 작업 T와 성능 측정 P에 대해 경험  E로 학습한 것이다. 

# 2.머신러닝을 사용하는 이유
스펨필터를 만든다면
1. 스펨 문자에 어떤 단어들이 주로 나타나는지 살펴본다.
2. 발견한 각 패턴을 감지하는 알고리즘을 작성하여 프로그램이 이런 패턴을 발견했을 때 그 메일을 스팸으로 분류하게 한다.
3. 충분한 성능이 나올때까지 1,2를 반복한다.

많은 경우의 수를 생각해야하고, 인력이 많이 든다.

반면\
머신러닝은 스팸메일에서 자주 나타나는 패턴을 감지하여 어떤 단어와 구절이 스팸 메일을 판단하는 데 좋은 기준인지 자동으로 학습한다.
그러므로 프로그램이 훨씬 짧아지고, 유지 보수가 쉬워지며, 정확도가 더 높다.

예전의 프로그래밍의 경우\
스팸 메일 발송자가 4U라는 단어가 스펨으로 분류된다는 것을 알아내고 For you로 고친다면 우린 새로운 알고리즘을 짜야한다. 즉 유지 보수를 위해영원히 새로운 규칙을 추가해야 한다.

반면\
머신러닝 기반의 스팸 필터는 바뀐 패턴도 자주 나타나는 것을 자동으로 인식하고, 스펨으로 분류한다.

또한\
머신러닝은 전통적인 방식으로는 너무 복잡하거나 알려진 알고리즘이 없는 문제의 경우 유용하다.

우리는 머신러닝을 통해 배울 수도 있다. 즉 머신러닝 알고리즘이 학습한 것을 조사할 수 있다. 예를 들어 스팸 필터를 만들었다면 스팸을 예측하는 데 가장 좋은 단어 및 단어의 조합이 무엇인지 확인할 수 있다. 이처럼 머신러닝 기술을 적용해서 대용량의 데이터를 분석하면 겉으로는 보이지 않던 패턴을 발견할 수 있는데 이를 **데이터 마이닝**이라고 한다.

# 3.머신러닝 시스템의 종류
머신러닝 시스템은 '학습하는 동안의 감독 형태나 정보량'에 따라 분류할 수 있다. 

### 3-1. 지도 학습

**지도 학습(supervised learning)**에는 알고리즘에 주입하는 훈련 데이터에 **레이블(label)**이라는 답이 포함된다.

**분류**가 전형적인 지도학습 작업이며, 스팸 필터가 좋은 예시이다.
또 다른 전형적인 작업은 **예측 변수(predictor variable)**라 부르는 **특성(feature)**을 사용해 중고차 가격 같은 타깃(target)수치를 예측하는 것이다. 이를 **회귀(Regression)**라고 부른다.

지도학습 알고리즘
- k-최근접 이웃(k-nearest neighbors)
- 선형 회귀(linear regression)
- 로지스틱 회귀(logistic regression)
- 서포트 벡터 머신(support vector machine)
- 결정 트리(decision tree)와 랜덤 포레스트(random forest)
- 신경망(neural networks)


### 3-2. 비지도 학습
**비지도 학습(unsupervised learning)**에는 훈련 데이터에 레이블이 없다, 시스템이 아무런 도움 없이 학습해야 한다.

비지도 학습 알고리즘
- 군집(clustering)
    - k-평균(k-means)
    - DBSCAN
    - 계층 군집 분석(hierachical cluster analysis, HCA)
    - 이상치 탐지(outlier detection)와 특이치 탐지(novelty detection) 
    - 원-클래스(one-class SVM)
    - 아이솔레이션 포레스트(isolation forest)
- 시각화(visualization)와 차원 축소(dimensionality reduction)
    - 주성분 분석(principal component analysis, PCA)
    - 커널(kernel PCA)
    - 지역적 선형 임베딩(locally-linear embedding, LLE)
    - t-SNE(t-distributed stochastic neighbor embedding)
- 연관 규칙 학습(association rule learning)
    - 어프라이어리(Apriori)
    - 이클렛(Eclat)

\
**시각화**알고리즘은 레이블이 없는 대규모의 고차원 데이터를 넣으면 도식화가 가능한 2D나 3D 표현을 만들어준다. 이런 알고리즘은 가능한 한 구조를 그대로 유지하려 하므로 데이터가 어떻게 조직되어 있는지 이해하고 예상치 못한 패턴을 발견할 수 있다.

비슷한 작업으로 너무 많은 정보를 일지 않으며 데이터를 간소화하려는 **차원 축소**가 있다. 이렇게 하는 한가지 방법은 상관관계가 있는 여러 특성을 하나로 합치는 것이다. \
예를 들어 차의 주행거리는 차의 연식과 강하게 연관되어 있으므로 차원 축소 알고리즘으로 두 특성을 차의 마모 정도를 나타내는 하나의 특성으로 합칠 수 있다. 이를 **특성 추출(feature extraction)**이라고 한다.

💡머신러닝 알고리즘에 데이터를 주입하기 전에 차원 축소를 사용하여 훈련 데이터의 차원을 줄이는 것이 유용할 때가 많다. 실행 속도가 훨씬 빨라지고 디스크와 메모리를 차지하는 공간도 줄고 경우에 따라 성능이 좋아지기도 한다.💡

\
**이상치 탐지**, 예를 들어 부정 거래를 막기 위해 이상한 신용카드 거래를 감지하고, 학습 알고리즘에 주입하기 전에 데이터셋에서 이상한 값을 자동으로 제거하는 것 등이다.\
시스템은 훈련하는 동안 대부분 정상 샘플을 만나 이를 인식하도록 훈련된다. 그 다음 샘플을 보고 정상 데이터인지 혹은 이상치인지 판단한다.

매우 비슷한 작업으로 **특이치 탐지**는 훈련 세트에 있는 모든 샘플과 달라 보이는 새로운 샘플을 탐지하는 것이 목적이다. 알고리즘으로 감지하고 싶은 모든 샘플을 제거한 "께끗한 훈련 세트"가 필요하다.

예를 들어 강아지 사진 수천장 중 1%가 치와와의 사진이라면 특이치 탐지 알고리즘이 치와와 사진을 새로운 특이한 것으로 처리하지 못한다, 반면 이상치 탐지 알고리즘은 치와와 사진을 매우 드물고 다른 강아지와 다르다고 인식하여 이상치로 분류한다.


**연관 규칙 학습**은 대량의 데이터에서 특성간의 흥미로운 관계를 찾는 것이다. \
예를 들어 바비큐 소스와 감자를 구매한 사람은 스테이크도 구매하는 경향을 찾아내서 가까이 진열 할 수도 있다.

### 3-3. 준지도 학습
데이터에 레이블을 다는 것은 시간, 비용이 많이 들기 때문에 레이블이 없는 샘플이 많고 레이블된 샘플은 적은 경우가 많다.\
**준지도 학습**(semisupervised learning)은 일부만 레이블이 있는 데이터를 다룰 수 있다. \
대부분의 준지도 학습 알고리즘은 지도 학습과 비지도 학습의 조합으로 이루어져 있다. 예를 들어 **심층 신뢰 신경망(deep belief network, DBN)**은 여러 겹으로 쌓은 **제한된 볼츠만 머신(restricted Boltzmann machine, RBM)이라 불리는 비지도 학습에 기초한다. RBM이 비지도 학습 방식으로 순차적으로 훈련된 다음 전체 시스템이 지도 학습 방식으로 세밀하게 조정된다.

### 3-4. 강화 학습
**강화학습(reinforcement learning)**은 학습하는 시스템을 **에이전트**라고 부르며 환경(environment)을 관찰해서 행동(action)을 실행하고 그 결과로 보상(reward) 또는 벌점(penalty)을 받는다. 시간이 지나면서 가장 큰 보상을 얻기 위해 **정책(policy)**이라고 부르는 최상의 전략을 스스로 학습한다. 정책은 주어진 상황에서 에이전트가 어떤 행동을 선택해야 할지 정의한다.

## 3-5.배치 학습과 온라인 학습
머신러닝 시스템은 입력 데이터의 스트림(stream)부터 점진적으로 학습할 수 있는지  여부로도 분류할 수 있다.

### 3-5-1. 배치 학습
**배치학습(batch learning)**에서는 시스템이 점진적으로 학습할 수 없다. 가용한 데이터를 모두 사용해 훈련시켜야 하는데 일반적으로 이 방식은 시간과 자원을 많이 소모하므로 오프라인에서 수행된다. 먼저 시스템을 훈련시키고 제품 시스템에 적용하면 더 이상의 학습 없이 실행된다. 즉 학습한 것을 단지 적용만 한다. 이를 **오프라인 학습(offline learning)**이라고 한다. 

배치 학습 시스템이 새로운 데이터에 대해 학습하려면 전체 데이터를 사용하여 시스템의 새로운 버전을 처음부터 다시 훈련해야 한다. 그런 뒤 이전 시스템을 중지시키고 새 시스템으로 교체한다. 


이러한 방식은 간단하고 잘 작동하지만 전체 데이터셋을 사용해 훈련하는 데 몇 시간이 소요될 수 있다. 이는 주식가격 같은 리얼타임 시스템에 적용해야 한다면 매우 힘들것이다. \
또한 전체 데이터셋을 사용해 훈련한다면 많은 컴퓨팅 자원이 필요하다.(큰 비용이 발생) 

이런 경우에는 점진적으로 학습할 수 있는 알고리즘을 사용해야 한다.

### 3-5-2. 온라인 학습
온라인 학습(online learning)에서는 데이터를 순차적으로 한 개씩 또는 미니배치(mini-batch)라 부르는 작은 묶음 단위로 주입하여 시스템을 훈련시킨다. 매 학습 단계가 빠르고 비용이 적게 들어 시스템은 데이터가 도착하는 대로 즉시 학습할 수 있다.

온라인 학습은 주식가격 같이 연속적으로 데이터를 받고 빠른 변화에 스스로 적응해야 하는 시스템에 적합하다. 컴퓨팅 자원이 제한된 경우에도 사용할 수 있다.
온라인 학습 시스템이 새로운 데이터 샘플을 학습하면 학습이 끝난 데이터는 버리면 되기 때문에, 많은 공간을 절약할 수 있다.

컴퓨터 한 대의 메인 메모리에 들어갈 수 없는 아주 큰 데이터셋을 학습하는 시스템에도 온라인 학습 알고리즘을 사용할 수 있다. 이를 **외부 메모리(out-of-core)**학습이라고 한다. 알고리즘이 데이터 일부를 읽어 들이고 훈련 단계를 수행하는 과정을 전체 데이터가 모두 적용될 때까지 반복하면 된다.

📢외부 메모리 학습은 컴퓨터의 주메모리에 들어갈 수 없는 대용량의 데이터를 다룰 수 있다. 보통 오프라인으로 실행된다. **점진적 학습(incremental learning)**이라고 생각하면 된다.📢

온라인 학습 시스템에서 중요한 파라미터는 변화하는 데이터에 얼마나 빠르게 적응할 것인지 이다. 이를 **학습률(learning rate)**이라고 한다. 학습률을 높게 하면 시스템이 데이터에 빠르게 적응하지만 예전 데이터를 금방 잊어버린다.
반대로 학습률이 낮으면 시스템의 관성이 커져서 느리게 학습된다. 하지만 새로운 데이터에 있는 잡음이나 대표성 없는 데이터 포인트에 덜 민감해진다. 

온라인 학습에서 가장 큰 문제점은 시스템에 나쁜 데이터가 주입되었을 때 시스템 성능이 점진적으로 감소한다는 점이다.

### 3-6-1. 사례 기반 학습
가장 간단한 형태의 학습은 단순히 기억하는 것이다. 예로 스팸 필터를 이런식으로 만든다면 사용자가 스팸이라고 지정한 메일과 동일한 모든 메일을 스팸으로 분류한다. 

이 방법 대신 스팸 메일과 매우 유사한 메일을 구분하도록 스팸 필터를 만들 수 있다.\
 이를 위해 두 메일 사이의 **유사도(similarity)**를 측정해야 한다. 두 메일 사이의 유사도 측정 방법은 공통으로 포함한 단어의 수를 세는 것이다. 스펨 메일과 공통으로 가지고 있는 단어의 수에 따라 분류한다. 

이를 **사례 기반 학습(instance-based learning)**이라고 한다. 시스템이 훈련 샘플을 기억함으로써 학습한다. 그리고 유사도 측정을 사용해 새로운 데이터와 학습한 샘플을 비교하는 식으로 일반화한다.

### 3-6-2. 모델 기반 학습
샘플로부터 일반화시키는 다른 방법은 이 샘플들의 모델을 만들어 **예측(prediction)**에 사용하는 것이다. \
이를 **모델 기반 학습(model-based learning)**이라고 한다.

image.png

위의 사진에서 새로운 샘플은 모델에 의해 세모로 분류된다.

모델 기반 학습의 예를 들어 삶의 만족도는 국가의 1인당 GDP에 따라 선형으로 증가하므로 선형 모델을 사용한다.
선형 모델은 두개의 모델 파라미터 세타(θ)0과 세타1을 갖는다. 이 모델 파라미터를 조정하여 선형 함수를 표현하는 모델을 얻을 수 있는데, 이 모델이 최상의 성능을 내도록하는 값을 알기 위해 측정 지표를 정하는데, 모델이 얼마나 좋은지 측정하는 **효용함수(utilty function)** 또는 **적합도 함수(fitness function)**를 정의하거나 얼마나 나쁜지 측정하는 **비용 함수(cost function)**를 정의할 수 있다 

# 4.머신러닝의 주요 도전 과제
주요 작업은 학습 알고리즘을 선택해 어떤 데이터에 훈련시키는 것이므로 문제가 될 수 있는 두 가지는 '나쁜 알고리즘'과 '나쁜 데이터'이다. 



### 4-1. 충분하지 않은 양의 훈련 데이터
대부분의 머신러닝 알고리즘이 잘 작동하려면 데이터가 많아야 한다. 아주 간단한 문제에서조차도 수천 개의 데이터가 필요하고 이미지나 음성 인식 같은 복잡한 문제라면 수백만 개가 필요할지도 모른다.


### 4-2. 대표성 없는 훈련 데이터
일반화가 잘되려면 우리가 일반화하기 원하는 새로운 사례를 훈련 데이터가 잘 대표하는 것이 중요하다. 하지만 이가 어려울 때가 많은데, 샘플이 작으면 **샘플링 잡음(sampling noise)**(즉 우연에 의한 대표성 없는 데이터)이 생기고, 매우 큰 샘플도 표본 추출 방법이 잘못되면 대표성을 띠지 못할 수 있다. 이를 **샘플링 편향(sampling bias)**라고 한다.

### 4-3. 낮은 품질의 데이터
훈련 데이터가 에러, 이상치, 잡음으로 가득하다면 머신러닝 시스템이 내재된 패턴을 찾기 어려워 잘 작동하지 않을 것이다. 이 때문에 훈련 데이터 정제가 중요하다. \
훈련 데이터 정제가 필요한 경우는
- 일부 샘플이 이상치란게 명확하면 간단히 그것들을 무시하거나 수동으로 잘못된 것을 고치는 것이 좋다. 
- 일부 샘플에 특성 몇 개가 빠져있다면(예를 들면 고객들 중 5%가 나이를 기록하지 않음) 이 특성을 모두 무시할지, 이 샘플을 무시할지, 빠진 값을 채울지, 또는 이 특성을 넣은 모델과 제외한 모델을 따로 훈련시킬 것인지 결정해야 한다.

### 4-4. 관련 없는 특성
훈련 데이터에 관련 없는 특성이 적고 관련 있는 특성이 충분해야 시스템이 학습할 수 있을 것이다. 훈련에 사용할 좋은 특성들을 찾는 과정을 **특성 공학(feature engineering)**이라고 하며 아래와 같은 작업이다.
- **특성 선택(feature selection)**: 가지고 있는 특성 중에서 훈련에 가장 유용한 특성을 선택
- **특성 추출(feature extraction)**: 특성을 결합하여 더 유용한 특성을 만든다. (차원 축소 알고리즘이 도움이 될 수 있다.)
- 새로운 데이터를 수집해, 새 특성을 만든다.

### 4-5. 훈련 데이터 과대적합
**과대적합(overfitting)**은 모델이 훈련 데이터에 너무 잘 맞지만 일반성이 떨어지는 것이다. \
훈련 세트에 잡음이 많거나 데이터셋이 너무 작으면 샘플링 잡음이 발생하므로 잡음이 섞인 패턴을 감지하게 된다. 감지한 패턴이 진짜인지 잡음 데이터로 인한 것인지 모델이 구분해낼 방법이 없기에 새로운 샘플에 일반화되지 못한다.

해결방법은 
- 파라미터 수가 적은 모델을 선택하거나 훈련 데이터에 있는 특성 수를 줄이거나, 모델에 제약을 가하여 단순화시킨다.
- 훈련 데이터를 더 많이 모으고, 잡음을 줄인다.

모델을 단순하게 하고 제약을 가하는 것을 **규제(regularization)**라고 한다. 

특성이 하나인 선형 모델은, 두개의 파라미터 절편과 기울기를 갖는데 이는 훈련 데이터에 모델을 맞추기 위한 두 개의 \
**자유도(degree of greedom)**를 학습 알고리즘에 부여한다. 모델은 직선의 절편과 기울기를 조절할 수 있다. 알고리즘이 기울기를 작은 값을 갖도록 유지시키면 학습 알고리즘이 자유도 1과 2사이의 적절한 어딘가에 위치하게 되고 자유도 1~2사이의 모델을 만든다. 즉 데이터에 완벽히 맞추는 것과 일반화를 위해 단순한 모델을 유지하는 것 사이의 올바른 균형을 찾는다.

위 사진에서 점선은 동그라미로 표현된 나라로 훈련한 모델이고, 파선은 동그라미와 사각형을 포함한 나라로 표현한 모델이며, 실선은 첫번째 모델과 같은 데이터에 규제를 적용해 만든 모델이다. 규제가 모델의 기울기를 더 작게 만들어, 동그라미에는 덜 맞지만 훈련 데이터셋에는 없는 새로운 샘플(사각형)에는 더 잘 일반화 된다. \


학습하는 동안 적용할 규제의 양은 **하이퍼파라미터(hyperparameter)**가 결정한다. 이는 학습 알고리즘의 파라미터이다. 그래서 학습 알고리즘으로부터 영향을 받지 않으며, 훈련 전에 미리 지정되고, 훈련하는 동안 상수로 남아있다. 


### 4-6. 훈련 데이터 과소적합
**과소적합(underfitting)**은 모델이 너무 단순해서 데이터의 내제된 구조를 학습하지 못할 때 일어난다.\
이 무제를 해결하는 기법
- 모델 파라미터가 더 많은 강력한 모델을 선택
- 학습 알고리즘에 더 좋은 트성을 제공한다.
- 모델의 제약을 줄인다.(규제 하이퍼파라미터를 감소시킨다.)



# 5.테스트와 검증
모델이 새로운 샘플에 얼마나 잘 일반화될지 아는 유일한 방법은 새로운 샘플에 실제로 적용해 보는 것이다. 

이에 **훈련세트**와 **테스트 세트**로 나눠서 훈련하고 테스트를 해보는 것이다. 새로운 샘플에 대한 오류 비율을\
**일반화 오차(generalization error)** 또는 **외부 샘플 오차(out-of-sample error)**라고 하며, 테스트 세트에서 모델을 평가함으로써 이 오차에 대한 추정값(estimation)을 얻는다. 이 값은 처음보는 샘플에 모델이 얼마나 잘 작동할지 알려준다.

### 5-1. 하이퍼파라미터 튜닝과 모델 선택
더 나은 모델과 하이퍼파라미터를 찾기 위해 일반화 오차를 테스트 세트에서 여러 번 측정하면 모델과 하이퍼파라미터가 테스트 세트에 최적화된 모델을 만든다. 이는 모델이 새로운 데이터에 잘 작동하지 않을 수 있다는 뜻이다. 

이 문제의 해결 방법은 **홀드아웃 검증(holdout validation)**이다. 훈련세트의 일부를 떼어내 여러 후보 모델을 평가하고 가장 좋은 하나를 택한다. 이 새로운 홀드아웃 세트를 **검증세트(validation set)**라고 한다. \
전체 데이터셋에서 검증세트를 뺀 데이터에서 다양한 하이퍼파라미터 값을 가진 여러 모델을 훈련한 뒤 검증 세트에서 가장 높은 성능을 내는 모델을 선택한다. 홀드아웃 검증 과정이 끝나면 이 최선의 모델을 검증 세트를 포함한 전체 훈련 세트에서 다시 훈련하여 최종 모델을 완성한다. 

하지만 검증세트가 너무 작으면 모델이 정확히 평가되지 않을 것이고 검증 세트가 너무 크면 남은 훈련 세트가 전체 훈련 세트보다 너무 작아진다. 이 문제를 해결하기 위해 작은 검증 세트를 여러 개 구성해 반복적인 **교차 검증(cross-validation)**을 수행하는 것이다. 검증 세트마다 나머지 데이터에서 훈련한 모델을 해당 검증 세트에서 평가한다.(모든 모델의 평가를 평균하면 훨씬 정확한 성능 측정이 가능하다) 하지만 시간이 검증 세트의 개수와 비례해 늘어난다.

### 5-2. 훈련-개발 세트
**훈련-개발 세트(train-dev set)**는 검증, 테스트 세트에 사용되는 데이터와 훈련 세트 사이에 데이터 불일치 위험이 있을 때 사용한다. 훈련 세트의 일부에서 모델을 훈련하고 훈련-개발 세트에서 평가할 때 나쁜 성능을 낸다면 아마 훈련 세트에 과대적합이 되었을 가능성이 높다. 훈련 세트와 훈련-개발 세트 양쪽에서 모두 잘 동작하지만 개발 세트에서 성능이 나쁘다면 훈련 데이터와 검증+테스트 데이터 사이에 데이터 불일치가 있을 가능성이 크다. 검증+테스트 데이터에 더 가깝게 훈련 데이터를 개선해야한다.

## The difference between Parameter and Hyper parameter

**파라미터**는 매개변수란 뜻이다. 파라미터는 모델 내부에서 결정되는 변수이다. 또한 그 값은 데이터로부터 결정된다. 예를 들어 한 반의 학생들의 키에 대한 정규분포를 그릴때, 평균과 표준편차 값을 구하게되는데 이 평균과 표준편차는 파라미터이다. 즉 파라미터는 데이터를 통해 구해지며 모델 내부적으로 결정되는 값이다.(사용자에 의해 조정되지 않는다.)

**하이퍼 파라미터**는 모델링할 때 사용자가 직접 세팅해주는 값을 뜻한다. learning rate나 SVMdml C 등이 하이퍼 파라미터이다. 하이퍼 파라미터는 정해진 최적의 값이 없다. 휴리스틱한 방법이나 경험 법칙(rules of thumb)에 의해 결정하는 경우가 많다. **베이지안 옵티미제이션**과 같이 자동으로 하이퍼 파라미터를 선택해주는 라이브러리가 있긴하다.
